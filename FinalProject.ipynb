```python
import sys
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

print("Enter file name: ")
file=open(input())
first=True
count=0
listoflists=[]
for line in file:
    line2=line.strip()
    if count<3:
        data=line2.split(",")
        count=count+1
    else:
        count=count+1
        data=line2.split(",")
        listoflists.append(data[5:])
        if count==47:
            break
newarray=np.array(listoflists)
converted_array=newarray.astype(float)
pineypoint=converted_array[0:11]
bishop=converted_array[11:22]
joebay=converted_array[22:33]
st_j_sound=converted_array[33:44]


print("Abbreviations: orthophosphates (OTP), phosphates (P), and Total Kjehldahl Nitrogen, (TKN)")
print()
## Mean Median and Stdev of all data ##
orth1, TP1, TKN1 =np.mean(converted_array, axis=0)
orth2, TP2, TKN2 =np.median(converted_array, axis=0)
orth3, TP3, TKN3=np.std(converted_array, axis=0)
print("From all four data sources combined: ")
print("Mean: OTP:", orth1, ", P:", TP1, ", TKN:", TKN1)
print("Median: OTP:", orth2, ", P:", TP2, ", TKN:",TKN2)
print("Standard Deviation: OTP:", orth3, ", P:", TP3, ", TKN:", TKN3)
print()
## Mean Median and Standard Deviation Piney Point##
orth1, TP1, TKN1 =np.mean(pineypoint, axis=0)
orth2, TP2, TKN2 =np.median(pineypoint, axis=0)
orth3, TP3, TKN3=np.std(pineypoint, axis=0)
print("Piney Point:")
print("Mean: OTP:", orth1, ", P:", TP1, ", TKN:", TKN1)
print("Median: OTP:", orth2, ", P:", TP2, ", TKN:",TKN2)
print("Standard Deviation: OTP:", orth3, ", P:", TP3, ", TKN:", TKN3)
print()
## Mean Median Stdev Bishop Harbor ##
orth1, TP1, TKN1 =np.mean(bishop, axis=0)
orth2, TP2, TKN2 =np.median(bishop, axis=0)
orth3, TP3, TKN3=np.std(bishop, axis=0)
print("Bishop Harbor:")
print("Mean: OTP:", orth1, ", P:", TP1, ", TKN:", TKN1)
print("Median: OTP:", orth2, ", P:", TP2, ", TKN:",TKN2)
print("Standard Deviation: OTP:", orth3, ", P:", TP3, ", TKN:", TKN3)
print()
## Mean Median Stdev Joe Bay ##
orth1, TP1, TKN1 =np.mean(joebay, axis=0)
orth2, TP2, TKN2 =np.median(joebay, axis=0)
orth3, TP3, TKN3=np.std(joebay, axis=0)
print("Joe Bay:")
print("Mean: OTP:", orth1, ", P:", TP1, ", TKN:", TKN1)
print("Median: OTP:", orth2, ", P:", TP2, ", TKN:",TKN2)
print("Standard Deviation: OTP:", orth3, ", P:", TP3, ", TKN:", TKN3)
print()
## Mean Median Stdev St. Joseph Sound ##
orth1, TP1, TKN1 =np.mean(st_j_sound, axis=0)
orth2, TP2, TKN2 =np.median(st_j_sound, axis=0)
orth3, TP3, TKN3=np.std(st_j_sound, axis=0)
print("St. Joseph Sound:")
print("Mean: OTP:", orth1, ", P:", TP1, ", TKN:", TKN1)
print("Median: OTP:", orth2, ", P:", TP2, ", TKN:",TKN2)
print("Standard Deviation: OTP:", orth3, ", P:", TP3, ", TKN:", TKN3)
print()

## Generating Histograms for total data ##
Orthop_tot= converted_array[:,[0]]
Phos_tot=converted_array[:,[1]]
TKN_tot=converted_array[:,[2]]
plt.hist(Orthop_tot, bins="auto", density=False)  
title="Histogram of Total Orthophosphates with 'auto' bins" 
plt.title(title)
plt.xlabel("Concentration (micrograms/L)")
plt.ylabel("Samples")
plt.show()
plt.hist(Phos_tot, bins="auto", density=False)  
title="Histogram of Total Phosphates with 'auto' bins" 
plt.title(title)
plt.xlabel("Concentration (micrograms/L)")
plt.ylabel("Samples")
plt.show()
plt.hist(TKN_tot, bins="auto", density=False)  
title="Histogram of TKN with 'auto' bins" 
plt.xlabel("Concentration (milligrams/L)")
plt.ylabel("Samples")
plt.title(title)
plt.show()

## Pearson and Spearman Correlation , Linreg, and Scatter Plots with Linear Regression lines##
orth1, TP1, TKN1 =np.mean(converted_array, axis=0)
orth2, TP2, TKN2 =np.median(converted_array, axis=0)
orth3, TP3, TKN3=np.std(converted_array, axis=0)
Op_t=[]
P_t=[]
TKN_t=[]
for line in Orthop_tot:
    Op_t.append(float(line))
for line in Phos_tot:
    P_t.append(float(line))
for line in TKN_tot:
    TKN_t.append(float(line))
x1, y1= stats.pearsonr(Op_t, P_t)
x11, y11=stats.spearmanr(Op_t, P_t)
x111=stats.linregress(Op_t, P_t)
slope, inter, rval, pval, std1 =stats.linregress(Op_t,P_t)
x2, y2=stats.pearsonr(Op_t, TKN_t)
x22, y22=stats.spearmanr(Op_t, TKN_t)
x222=stats.linregress(Op_t, TKN_t)
x3, y3=stats.pearsonr(P_t, TKN_t)
x33, y33=stats.spearmanr(P_t, TKN_t)
x333=stats.linregress(P_t, TKN_t)
print("Between (OTP) and (P):")
print("Pearson:", x1, " Spearman:", x11)
print(x111)
fig, ax=plt.subplots()
ax.scatter(Op_t, P_t, c="grey")
ax.set_xlabel("Concentration OTP (micrograms/L)")
ax.set_ylabel("Concentration P (micrograms/L)")
plt.title("OTP vs P")
plt.plot(Orthop_tot, Orthop_tot*slope + inter, color='red')
plt.show()
print()
print("Between (OTP) and (TKN):")
print("Pearson:",x2, " Spearman:", x22) 
print(x222)
slope, inter, rval, pval, std1 =stats.linregress(Op_t,TKN_t)
fig, ax=plt.subplots()
ax.scatter(Op_t, TKN_t, c="grey")
ax.set_xlabel("Concentration OTP (micrograms/L)")
ax.set_ylabel("Concentration TKN (mg/L)")
plt.title("OTP vs TKN")
plt.plot(Orthop_tot, Orthop_tot*slope + inter, color='red')
plt.show()
print()
print("Between (P) and (TKN):")
print("Pearson:", x3, " Spearman:", x33)
print(x333)
slope, inter, rval, pval, std1 =stats.linregress(P_t,TKN_t)
fig, ax=plt.subplots()
ax.scatter(P_t, TKN_t, c="grey")
ax.set_xlabel("Concentration P (micrograms/L)")
ax.set_ylabel("Concentration TKN (mg/L)")
plt.title("P vs TKN")
plt.plot(Phos_tot, Phos_tot*slope + inter, color='red')
plt.show()

## PRINCIPLE COMPONENT ANALYSIS ##

pca=PCA()
principal_components = pca.fit_transform(converted_array)


fig, ax=plt.subplots()
PCX=0
PCY=1

ax.scatter(principal_components[:, PCX], principal_components[:, PCY], c="red")
print("Explained Variances: ", pca.explained_variance_ratio_)
label_X = "PCX (" + str(np.round(pca.explained_variance_ratio_[PCX]*100, decimals=2)) + "%)"
label_Y = "PCY (" + str(np.round(pca.explained_variance_ratio_[PCY]*100, decimals=2)) + "%)"
ax.scatter(principal_components[0:10 , PCX], principal_components[0:10 , PCY], c="green")  
ax.scatter(principal_components[11:21 , PCX], principal_components[11:21 , PCY], c="blue")  
ax.scatter(principal_components[22:32 , PCX], principal_components[22:32 , PCY], c="orange")  
ax.scatter(principal_components[33:43 , PCX], principal_components[33:43 , PCY], c="red")
ax.set_xlabel(label_X)
ax.set_ylabel(label_Y)
plt.title("Principle Component Analysis Graphically")
plt.show()
fig, ax=plt.subplots()

plt.bar(range(3), pca.explained_variance_ratio_)
ax.set_xlabel("Principal Component")
ax.set_ylabel("Explained Variance Ratio")
plt.title("Explained Variance by Principal Component")

plt.show()

```
